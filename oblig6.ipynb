```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline

# f2c_shortcut_plot

# HUsk kjoreeksempler

Tmin = -20
Tmax = 120
N = 140
F = np.linspace(Tmin,Tmax,N)

C_quick = np.zeros(N)
C_quick = (F-30)/2
C_exact = np.zeros(N)
C_exact = (F-32)*5/9

line_quick, = plt.plot(F,C_quick, label = 'Quick method')
line_exact, = plt.plot(F,C_exact, label = 'Exact method')
plt.legend(handles = [line_quick, line_exact], loc = 'best')
plt.ylabel('Celsius')
plt.xlabel('Fahrenhet')
plt.title('Fahrenheit to Celsius conversion')
```

```python
# plot_wavepacket.py
from math import exp,sin,pi
# import matplot.., numpy, seaborn

def wave(x,t):
    return exp(-(x-3*t)**2)*sin(3*pi*(x-t))

N = 100
x = np.linspace(-4,4,N)
t = 0

f = np.zeros(N)
for i in range(N):
    f[i] = wave(x[i],t)
    
plt.plot(x,f)
plt.ylabel('wave')
plt.xlabel('x')
plt.title('Wave locolized in space for t = 0')
```

```python
# water_wave_velocity.py

# import matplotl, seaborn, pi, numpy
from math import sqrt,tanh

def wave_speed(lam):
    g = 9.81
    s = 7.9*10**(-2)
    rho = 1000
    h = 50
    return sqrt(g*lam/(2*pi)*(1+s*4*pi**2/(rho*g*lam**2))*tanh(2*pi*h/lam))

N = 1000
lam_small = np.linspace(0.001,0.1,N)
lam_large = np.linspace(1000,2000,N)

c_small = np.zeros(N)
c_large = np.zeros(N)
for i in range(N):
    c_small[i] = wave_speed(lam_small[i])
    c_large[i] = wave_speed(lam_large[i])

fig = plt.figure()
sub1 = fig.add_subplot(211)
sub1.plot(lam_small,c_small)
sub2 = fig.add_subplot(212)
sub2.plot(lam_large,c_large)
sub2.set_xlabel('$\lambda$ [m]')
fig.text(0.03, 0.5, 'Wave speed, c [m/s]', ha='center', va='center', rotation='vertical')
```

```python
# sinesum1_plot.py

# import sin, pi, numpy

N = 20
T = 2*pi
t = np.linspace(0,T,N)

def S(t,n,T):
    summ = 0
    T = 2*pi
    for i in range(n):
        summ += 1/(2*i-1)*sin((2*(2*i-1)*pi*t)/T)
    return 4/pi*summ

def f(t,T):
    if t < T/2:
        func = -1
    elif t == T/2:
        func = 0
    else:
        func = 1
    return func

f_exact = np.zeros(N)
S1 = np.zeros(N)
S3 = np.zeros(N)
S20 = np.zeros(N)
S200 = np.zeros(N)

for i in range(N):
    f_exact[i] = f(i,T)
    S1[i] = S(i,1,T)
    S3[i] = S(i,3,T)
    S20[i] = S(i,20,T)
    S200[i] = S(i,200,T)
print f_exact
    
linef, = plt.plot(t,f_exact, label='exaxt f')
lineS1, = plt.plot(t,S1, label = 'S(t;1)')
lineS3, = plt.plot(t,S3, label = 'S(t;3)')
lineS20, = plt.plot(t,S20, label = 'S(t;20)')
lineS200, = plt.plot(t,S200, label = 'S(t;200)')
plt.legend(handles=[linef, lineS1, lineS3, lineS20, lineS200], loc = 'best')
```
